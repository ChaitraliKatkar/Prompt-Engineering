### Generating Q&A Pairs from Personal Finance Dataset

```python
# Generate Q&A pairs from the dataset
def generate_qna_from_finance(df):
    qna_data = []
    for _, row in df.iterrows():
        if row['Income/Expense'] == 'Income':
            question = f"How much income did I receive from {row['Sub category']} on {row['Date / Time']}?"
            answer = f"You received an income of ${row['Debit/Credit']} from {row['Sub category']} on {row['Date / Time']}."
        else:
            question = f"How much did I spend on {row['Category']} ({row['Sub category']}) on {row['Date / Time']}?"
            answer = f"You spent ${row['Debit/Credit']} on {row['Category']} ({row['Sub category']}) on {row['Date / Time']}."
        qna_data.append({"question": question, "answer": answer})
    return qna_data

# Generate the Q&A pairs
qna_pairs = generate_qna_from_finance(df)

# Convert to DataFrame for better visibility
qna_df = pd.DataFrame(qna_pairs)

# Display the first few Q&A pairs
qna_df.head()

output_file = "qna_pairs.json"
with open(output_file, "w") as file:
    json.dump(qna_pairs, file, indent=4)

print(f"Q&A pairs have been saved to {output_file}")
```

### Fine-Tuning T5 Model on Personal Finance Q&A Dataset

```python
import json
from datasets import Dataset
from transformers import T5Tokenizer, T5ForConditionalGeneration, Trainer, TrainingArguments

# Define the path to your Q&A JSON file
qna_file_path = "qna_pairs.json"  # Update this path as needed

# Load Q&A pairs
with open(qna_file_path, "r") as f:
    qna_data = json.load(f)

# Convert to Hugging Face Dataset
train_data = Dataset.from_list(qna_data)

# Load tokenizer and model
tokenizer = T5Tokenizer.from_pretrained("t5-small")
model = T5ForConditionalGeneration.from_pretrained("t5-small")

# Tokenize the dataset
def preprocess_function(sample):
    input_text = "question: " + sample["question"]
    target_text = "answer: " + sample["answer"]
    return {
        "input_ids": tokenizer(input_text, max_length=128, truncation=True, padding="max_length")['input_ids'],
        "labels": tokenizer(target_text, max_length=64, truncation=True, padding="max_length")['input_ids']
    }

train_data = train_data.map(preprocess_function)

# Set training arguments
training_args = TrainingArguments(
    output_dir="./fine_tuned_t5",
    per_device_train_batch_size=4,
    num_train_epochs=1,  # Adjust epochs as needed
    logging_dir="./logs",
    save_strategy="epoch"
)

# Trainer
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_data
)

# Train the model
trainer.train()

# Save the fine-tuned model
model.save_pretrained("./fine_tuned_t5")
print("Fine-tuning completed and model saved!")
```

### Validating the Format of Q&A JSON File

```python
import json

# Path to the Q&A file
file_path = "qna_pairs.json"

try:
    # Load the JSON file
    with open(file_path, "r") as f:
        data = json.load(f)

    # Validation checks
    if not isinstance(data, list):
        raise ValueError("The JSON file must contain a list of Q&A pairs.")

    for idx, item in enumerate(data):
        if not isinstance(item, dict):
            raise ValueError(f"Item at index {idx} is not a dictionary: {item}")
        
        if "question" not in item or not isinstance(item["question"], str):
            raise ValueError(f"Missing or invalid 'question' at index {idx}: {item}")
        
        if "answer" not in item or not isinstance(item["answer"], str):
            raise ValueError(f"Missing or invalid 'answer' at index {idx}: {item}")

    print("✅ Q&A JSON file format is valid.")

except json.JSONDecodeError as e:
    print(f"❌ Invalid JSON file: {e}")
except ValueError as e:
    print(f"❌ Validation error: {e}")
except Exception as e:
    print(f"❌ Unexpected error: {e}")
```

### Uploading Q&A Pairs to Weaviate with Custom Embeddings

```python

import requests
import json
from sentence_transformers import SentenceTransformer

# ----------------------------
# CONFIGURATION
# ----------------------------
WEAVIATE_URL = "http://localhost:9090"
COLLECTION_NAME = "FinanceQA"
SCHEMA_ENDPOINT = f"{WEAVIATE_URL}/v1/schema"

# Load Q&A pairs from the file
file_path = "qna_pairs.json"
try:
    with open(file_path, "r") as f:
        qna_data = json.load(f)
except Exception as e:
    print(f"Error loading Q&A pairs file: {e}")
    exit(1)

# ----------------------------
# DELETE EXISTING COLLECTION
# ----------------------------
try:
    response = requests.delete(f"{SCHEMA_ENDPOINT}/{COLLECTION_NAME}")
    if response.status_code == 204:
        print(f"Existing collection '{COLLECTION_NAME}' deleted.")
    else:
        print(f"No existing collection '{COLLECTION_NAME}' to delete.")
except Exception as e:
    print(f"Error deleting collection: {e}")
    exit(1)

# ----------------------------
# CREATE COLLECTION SCHEMA
# ----------------------------
schema = {
    "class": COLLECTION_NAME,
    "description": "Q&A pairs for personal finance chatbot",
    "vectorizer": "none",  # Disable default vectorizer for custom embeddings
    "properties": [
        {"name": "question", "dataType": ["text"]},
        {"name": "answer", "dataType": ["text"]},
    ],
}

try:
    response = requests.post(SCHEMA_ENDPOINT, json=schema)
    if response.status_code == 200:
        print(f"Collection '{COLLECTION_NAME}' created successfully.")
    else:
        print(f"Error creating collection: {response.text}")
        exit(1)
except Exception as e:
    print(f"Error creating collection: {e}")
    exit(1)

# ----------------------------
# INITIALIZE EMBEDDING MODEL
# ----------------------------
try:
    embed_model = SentenceTransformer("all-MiniLM-L6-v2")
    print("SentenceTransformer model loaded successfully.")
except Exception as e:
    print(f"Error loading embedding model: {e}")
    exit(1)

# ----------------------------
# INSERT DATA INTO COLLECTION
# ----------------------------
data_endpoint = f"{WEAVIATE_URL}/v1/objects"
headers = {"Content-Type": "application/json"}

for item in qna_data:
    question = item.get("question", "")
    answer = item.get("answer", "")

    if not question or not answer:
        print(f"Skipping invalid Q&A pair: {item}")
        continue

    # Generate embedding vector
    vector = embed_model.encode(question).tolist()

    # Create object payload
    weaviate_object = {
        "class": COLLECTION_NAME,
        "properties": {
            "question": question,
            "answer": answer,
        },
        "vector": vector,
    }

    # Insert object into Weaviate
    try:
        response = requests.post(data_endpoint, json=weaviate_object, headers=headers)
        if response.status_code != 200:
            print(f"Error inserting object: {response.text}")
    except Exception as e:
        print(f"Error inserting data: {e}")

print("Q&A pairs uploaded to Weaviate successfully!")

```

### Querying Weaviate for Relevant Q&A Pairs Using Semantic Search(RAG)

```python
import requests
from sentence_transformers import SentenceTransformer

# ----------------------------
# CONFIGURATION
# ----------------------------
WEAVIATE_URL = "http://localhost:9090"
COLLECTION_NAME = "FinanceQA"
QUERY_ENDPOINT = f"{WEAVIATE_URL}/v1/graphql"

# Initialize SentenceTransformer model
embed_model = SentenceTransformer("all-MiniLM-L6-v2")

def query_weaviate(user_query, k=3):
    # Generate embedding for the user query
    vector = embed_model.encode(user_query).tolist()

    # Define GraphQL query
    graphql_query = {
        "query": f"""
        {{
          Get {{
            {COLLECTION_NAME} (
              nearVector: {{ vector: {vector} }}
              limit: {k}
            ) {{
              question
              answer
            }}
          }}
        }}
        """
    }

    # Send request to Weaviate
    headers = {"Content-Type": "application/json"}
    response = requests.post(QUERY_ENDPOINT, json=graphql_query, headers=headers)

    # Process response
    if response.status_code == 200:
        results = response.json().get("data", {}).get("Get", {}).get(COLLECTION_NAME, [])
        if results:
            print("\nRetrieved Results:")
            for idx, item in enumerate(results):
                print(f"\nResult {idx + 1}:")
                print(f"Question: {item['question']}")
                print(f"Answer: {item['answer']}")
        else:
            print("No relevant results found.")
    else:
        print(f"Error: {response.text}")

# Example Query
if __name__ == "__main__":
    user_input = input("Enter your question: ")
    query_weaviate(user_input)
```

### Spending Query Bot: Retrieve and Summarize Expenses Using Weaviate and GPT-2

```python

import requests
from sentence_transformers import SentenceTransformer
from transformers import pipeline
import re  # Importing regular expressions to extract the amount

# ----------------------------
# CONFIGURATION
# ----------------------------
WEAVIATE_URL = "http://localhost:9090"
COLLECTION_NAME = "FinanceQA"
QUERY_ENDPOINT = f"{WEAVIATE_URL}/v1/graphql"

# Initialize embedding model and language model
embed_model = SentenceTransformer("all-MiniLM-L6-v2")
llm = pipeline("text-generation", model="gpt2")  # Use GPT-2 as a supported model


def retrieve_from_weaviate(query, k=3):
    """
    Retrieve relevant results from Weaviate using the query and embedding.
    """
    vector = embed_model.encode(query).tolist()
    graphql_query = {
        "query": f"""
        {{
          Get {{
            {COLLECTION_NAME} (
              nearVector: {{ vector: {vector} }}
              limit: {k}
            ) {{
              question
              answer
            }}
          }}
        }}
        """
    }
    headers = {"Content-Type": "application/json"}
    try:
        response = requests.post(QUERY_ENDPOINT, json=graphql_query, headers=headers)
        response.raise_for_status()
        results = response.json().get("data", {}).get("Get", {}).get(COLLECTION_NAME, [])
        return results
    except requests.RequestException as e:
        print(f"Error connecting to Weaviate: {e}")
        return []


def generate_response(user_query):
    """
    Generate a response by querying Weaviate and calculating total spending on Food (Snacks).
    """
    # Retrieve relevant data from Weaviate
    retrieved_data = retrieve_from_weaviate(user_query)
    
    # Filter for Food (Snacks) related entries
    snacks_data = [item for item in retrieved_data if "Food (Snacks)" in item.get("question", "")]

    # If there are relevant snack entries, calculate total spending
    if snacks_data:
        total_spending = 0
        for item in snacks_data:
            # Use regular expressions to find the dollar amount in the answer
            match = re.search(r"\$([\d\.]+)", item.get("answer", ""))
            if match:
                total_spending += float(match.group(1))  # Add the extracted amount to the total
        response_text = f"The total amount spent on Food (Snacks) is ${total_spending:.2f}."
    else:
        response_text = "I couldn't find any records for spending on Food (Snacks)."

    return response_text


# ----------------------------
# MAIN INTERACTION LOOP
# ----------------------------
if __name__ == "__main__":
    print("Welcome to the Spending Query Bot!")
    print("Type 'exit' to stop the conversation.\n")
    while True:
        # Get user input
        user_input = input("Enter your question: ").strip()
        if user_input.lower() == "exit":
            print("Goodbye!")
            break

        # Generate and print the response
        response = generate_response(user_input)
        print("\nBot Response:")
        print(response)
        print("\n" + "-" * 50)
```

```python
import weaviate
import json
from sentence_transformers import SentenceTransformer
from weaviate.connect import ConnectionParams

# ----------------------------
# CONFIGURATION
# ----------------------------
WEAVIATE_URL = "localhost"
COLLECTION_NAME = "FinanceQA"

# Initialize Weaviate Client with HTTP and gRPC configurations
connection_params = ConnectionParams(
    http={"host": WEAVIATE_URL, "port": 9090, "secure": False},
    grpc={"host": WEAVIATE_URL, "port": 50051, "secure": False}  # gRPC connection
)
client = weaviate.WeaviateClient(connection_params=connection_params)

# Explicitly connect to Weaviate
try:
    client.connect()
    print("Connected to Weaviate successfully!")
except Exception as e:
    print(f"Error connecting to Weaviate: {e}")
    exit(1)

# Load Q&A pairs from the JSON file
with open("qna_pairs.json", "r") as f:
    qna_data = json.load(f)

# Define the schema for the collection
schema = {
    "class": COLLECTION_NAME,
    "vectorizer": "text2vec-transformers",  # Ensure transformer-based vectorization
    "properties": [
        {"name": "question", "dataType": ["text"]},
        {"name": "answer", "dataType": ["text"]}
    ]
}

# Create the schema if it doesn't exist
try:
    if COLLECTION_NAME not in client.collections.list_all():
        client.collections.create(schema=schema)
        print(f"Collection '{COLLECTION_NAME}' created successfully.")
except Exception as e:
    print(f"Error managing schema: {e}")
    exit(1)

# Initialize SentenceTransformer for embeddings
embed_model = SentenceTransformer("all-MiniLM-L6-v2")

# Upload Q&A pairs to Weaviate
try:
    collection = client.collections.get(COLLECTION_NAME)

    for item in qna_data:
        # Generate vector embeddings
        vector = embed_model.encode(item["question"]).tolist()

        # Insert data into the collection with custom vectors
        collection.data.insert(
            properties={
                "question": item["question"],
                "answer": item["answer"],
            },
            vector=vector
        )
    print("Q&A pairs uploaded successfully to Weaviate!")
except Exception as e:
    print(f"Error uploading data to Weaviate: {e}")
```

```python
import requests
from sentence_transformers import SentenceTransformer

# ----------------------------
# CONFIGURATION
# ----------------------------
WEAVIATE_URL = "http://localhost:9090"  # Replace with your Weaviate URL
COLLECTION_NAME = "FinanceQA"           # Weaviate Collection name
QUERY_ENDPOINT = f"{WEAVIATE_URL}/v1/graphql"

# Initialize SentenceTransformer model
embed_model = SentenceTransformer("all-MiniLM-L6-v2")

# Function to retrieve results from Weaviate
def fetch_results_from_weaviate(query, certainty_threshold=0.5, limit=5):
    # Generate embeddings for the query
    vector = embed_model.encode(query).tolist()
    
    # GraphQL Query
    graphql_query = {
        "query": f"""
        {{
          Get {{
            {COLLECTION_NAME} (
              nearVector: {{
                vector: {vector},
                certainty: {certainty_threshold}
              }}
              limit: {limit}
            ) {{
              question
              answer
            }}
          }}
        }}
        """
    }
    headers = {"Content-Type": "application/json"}

    try:
        # Send request to Weaviate
        response = requests.post(QUERY_ENDPOINT, json=graphql_query, headers=headers)
        response.raise_for_status()

        # Parse response
        results = response.json().get("data", {}).get("Get", {}).get(COLLECTION_NAME, [])
        return results
    except requests.exceptions.RequestException as e:
        print(f"Error: {e}")
        return []
    except Exception as e:
        print(f"Unexpected error occurred: {e}")
        return []

# Run the script
if __name__ == "__main__":
    print("Ask me anything about your expenses and income!\n")
    user_input = input("Enter your question: ").strip()

    if user_input:
        print("\nFetching results...\n")
        results = fetch_results_from_weaviate(user_input, certainty_threshold=0.4)

        # Display the results
        if results:
            print("Results Found:\n")
            for idx, result in enumerate(results, start=1):
                print(f"Record {idx}:")
                print(f"🔹 **Question:** {result.get('question', 'N/A')}")
                print(f"🔹 **Answer:** {result.get('answer', 'N/A')}")
        else:
            print("\nNo relevant results found for your query.")
    else:
        print("No query entered. Please provide a valid question.")
```
